# 04-02 : Inspecting Container Settings

```python
import os
```

## 1. Configuration 

```python
# set the mlrun api path
mlrun_api_ip = !(hostname -I | awk '{print $1}')
mlrun_api_ip = mlrun_api_ip[0]
mlrun_api_port = 30070

os.environ["MLRUN_DBPATH"] = f"http://{mlrun_api_ip}:{mlrun_api_port}"
print(f"MLRUN_DBPATH: {os.environ['MLRUN_DBPATH']}")
```

```python
import mlrun
```

## 2. Create Project

```python
project = mlrun.get_or_create_project(
    name='hello-project',
    context='./',) 
```

## 3. Create a new function

```python
inspector_func = project.set_function(
    name="env-inspector",
    func="inspector.py",  # The script file
    handler="inspect_environment",  # The function to call in the script
    image="mlrun/mlrun",  # Using a standard MLRun image
    kind="job",
)
```

```python
# Add a custom environment variable to see if it appears
inspector_func.set_env("MY_TEST_VARIABLE", "ThisIsATestValue123")
inspector_func.set_env("ANOTHER_CUSTOM_ENV", "HelloFromMLRunSetup")
```

```python
print("Running the environment inspector function...")
run_uid = project.run_function(
    inspector_func,
    local=True, # Set to False if you have a full MLRun cluster and want to run it there
    auto_build=True,
    # watch=True # Uncomment to stream logs to console
).uid()

print(f"Inspector function run submitted. Run UID: {run_uid}")
print("Check the MLRun UI or use 'mlrun get run --uid <UID>' for logs and artifacts.")
print(f"Example CLI: mlrun get run --uid {run_uid} --project {project_name}")
print(f"Or view logs: mlrun logs --uid {run_uid} --project {project_name}")
```
